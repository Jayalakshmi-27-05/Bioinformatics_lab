```python
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML

fasta_sequences = { "SeqA": "ATGCGTACGTAC","SeqB": "ATGCGTAGCTAC"}
def simple_sequence_comparison(seq1, seq2, match_score=1, mismatch_penalty=-1, gap_penalty=-2):
    score = 0
    alignment_seq1 = ""
    alignment_seq2 = ""
    i, j = 0, 0

    while i < len(seq1) and j < len(seq2):
        if seq1[i] == seq2[j]:
            score += match_score
        else:
            score += mismatch_penalty
        alignment_seq1 += seq1[i]
        alignment_seq2 += seq2[j]
        i += 1
        j += 1

    while i < len(seq1):
        score += gap_penalty
        alignment_seq1 += seq1[i]
        alignment_seq2 += '-'
        i += 1
    while j < len(seq2):
        score += gap_penalty
        alignment_seq1 += '-'
        alignment_seq2 += seq2[j]
        j += 1

    return score, alignment_seq1, alignment_seq2
seq1 = fasta_sequences["SeqA"]
seq2 = fasta_sequences["SeqB"]
score, aligned_seq1, aligned_seq2 = simple_sequence_comparison(seq1, seq2)

print("Simple Comparison Results:")
print(f"Sequence 1: {seq1}")
print(f"Sequence 2: {seq2}")
print(f"Alignment Score: {score}")
print(f"Aligned Sequence 1: {aligned_seq1}")
print(f"Aligned Sequence 2: {aligned_seq2}")

def perform_blast(sequence_ids):
    for seq_id in sequence_ids:
        print(f"\n Performing BLAST for: {seq_id}")
        result_handle = NCBIWWW.qblast("blastn", "nt", seq_id)
        blast_records = NCBIXML.parse(result_handle)

        for blast_record in blast_records:
            print(f"Query: {blast_record.query}")
            for alignment in blast_record.alignments:
                for hsp in alignment.hsps:
                    print(f"Alignment: {alignment.title}")
                    print(f"Length: {alignment.length}")
                    print(f"E-value: {hsp.expect}")
                    print(f"Score: {hsp.score}, Identities: {hsp.identities}, Gaps: {hsp.gaps}")
                    print(f"Query: {hsp.query[:75]}...")
                    print(f"Match: {hsp.match[:75]}...")
                    print(f"Sbjct: {hsp.sbjct[:75]}...")
                    print("-" * 60)
        result_handle.close()
sequence_ids = ["PU773469.1", "PU773470.1"]
perform_blast(sequence_ids)
```
